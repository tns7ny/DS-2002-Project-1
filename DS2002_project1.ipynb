# 2. Modify

Modify the number of columns from the source to the destination. We choose to reduce the number of columns

```python
# Import pandas
import pandas as pd

# Load in Top Beauty & Cosmetics Products Worldwide 2024 (CSV) Data
try:
  top_cosmetics_wrldwide_2024 = pd.read_csv('/content/most_used_beauty_cosmetics_products_extended.csv', sep=',', encoding='latin1')
  print("CSV file has been read")
except FileNotFoundError as e:
  print("CSV file not found, please check that the file has been uploaded")

# Load in Cosmetic Brand Products Dataset (JSON) Data
try:
  cosmetic_brand_products = pd.read_json('/content/makeup_data.json')
  print("JSON file has been read")
except FileNotFoundError as e:
  print("JSON file not found, please check that the file has been uploaded")
```

**Modifying the Number of Columns**


```python
# Reduce number of columns in Top Cosmetics
top_cosmetics_wrldwide_2024.drop('Usage_Frequency', axis=1, inplace=True)
top_cosmetics_wrldwide_2024.drop('Packaging_Type', axis=1, inplace=True)

top_cosmetics_wrldwide_2024.head()
```

```python
# Reduce number of columns in Cosmetic Brands
cosmetic_brand_products.drop('rating', axis=1, inplace=True)
cosmetic_brand_products.drop('product_colors', axis=1, inplace=True)
cosmetic_brand_products.drop('tag_list', axis=1, inplace=True)

cosmetic_brand_products.head()
```

# 3. Convert

 Convert the general format and data structure of the data source (from
JSON to CSV, from CSV to JSON, from JSON into a SQL database table,
etc. I want the option to convert any source to any target. So, if I get a
CSV as an input, I want the user to choose an output)

Get user input to choose an output (JSON or CSV or SQL database table)


```python
output = input()
```

Convert to JSON

```python
if output == 'JSON' or output == 'json':
  cosmetic_brand_products.to_json('cosmetic_brand_products.json', index=False)
  top_cosmetics_wrldwide_2024.to_json('top_cosmetics_worldwide_2024.json', index=False)
```

Convert to CSV

```python
if output == 'CSV' or output == 'csv':
  cosmetic_brand_products.to_csv('cosmetic_brand_products.csv', index=False)
  top_cosmetics_wrldwide_2024.to_csv('top_cosmetics_worldwide_2024.csv', index=False)
```

Convert to SQL database table

```python
from sqlalchemy import create_engine
engine1 = create_engine('sqlite:///sql_cosmetic_brand_products.db', echo=False)
engine2 = create_engine('sqlite:///sql_top_cosmetics_worldwide_2024.db', echo=False)

if output == 'SQL database table' or output == 'SQL':
  cosmetic_brand_products.to_sql('cosmetic_brand_products', con=engine1, if_exists='replace')
  top_cosmetics_wrldwide_2024.to_sql('top_cosmetics_worldwide_2024', con=engine2, if_exists='replace')
```

```python
if output != 'JSON' and output != 'json' and output != 'CSV'and output != 'csv' and output != 'SQL' and output != 'SQL database table':
  print("Please enter one of the following outputs: JSON, CSV, or SQL")
```

# 4. Store

The converted (new) file should be written to disk (local file) or written to a SQL database

```python
cosmetic_brand_products.to_sql('cosmetic_brand_products', con=engine1, if_exists='replace')
top_cosmetics_wrldwide_2024.to_sql('top_cosmetics_worldwide_2024', con=engine2, if_exists='replace')
```
